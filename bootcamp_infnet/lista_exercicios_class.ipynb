1) Dado a lista, abaixo, de temperaturas da cidade de Boulder em Fahrenheit (F):

 1.1) Crie uma lista convertendo para Celsius (C). Para converter use a fórmula C = (F-32)*(5/9); Print os valores na tela. 
 
 1.2) Crie uma função que receba a lista de temperaturas em Fahrenheit e retorna uma lista em graus Celsius. Print a lista na tela. 

 1.3) **(Desafio)** Generalize a função criada em 1.2 usando *args para que a função receba  uma quantidade variável de argumentos.
 

 F = [47,49,57,64,72,83,89,87,79,67,55,47]; (Temperaturas em F da cidade de Boulder)


```python
def f_to_c(*temp_in_f_list):
    temp_in_c_list = [(f - 32)*(5/9) for f in temp_in_f_list]
    return temp_in_c_list
    
fahrenheit = [47,49,57,64,72,83,89,87,79,67,55,47]

c = f_to_c(*fahrenheit)
print(c)
```

(a) Make a class called User. 

(b) Create two attributes called first_name and last_name, and then create several other attributes that are typically stored in a user profile. 

(c) Make a method called describe_user() that prints a summary of the user’s information. 

(d) Make another method called greet_user() that prints a personalized greeting to the user.

```python
class User:
    def __init__(self,first_name,last_name,email,user_age):
        self.first_name = first_name
        self.last_name = last_name
        self.email = email
        self.user_age = user_age
        self.login_attempts = 0
    
    def describe_user(self):
        print(f'''
        Primeiro nome: {self.first_name}
        Sobrenome: {self.last_name}
        E-mail: {self.email}
        Idade: {self.user_age}
        ''')
    
    def greet_user(self):
        print(f"Seja bem-vindo, {self.first_name}!")
    
    def increment_login_attempts(self):
        self.login_attempts += 1
    
    def reset_login_attempts(self):
        self.login_attempts = 0
        
```

```python
my_user = User('Bernardo','Zottolo','bernardo.zottolo@hotmail.com',23)
my_user.describe_user()
```

```python
my_user.greet_user()
```

3) Usando o e-book 'Alice’s Adventures in Wonderland' contido em Alice.txt: 

(a) Faça um programa que retorne o número total de palavras contidas no e-book; 

(b) Um programa que retorne um dicionário contendo a frequência de cada palavra. Por exemplo: se a palavra 'guerra' aparece 10 vezes em um texto que contém 50 palavras, então a frequência da palavra 'guerra' é de 20%. 

(c) Retorne as 20 palavras mais frequentes do e-book;

```python
file1 = open(FileName, 'r')
todas_as_palavras = []
for i in file1:
    for j in i.split():
        palavra = ''
        for caractere in j:
            if caractere.isalpha() and caractere != "'": #Vai verificar caractere por caractere das palavras e remover os caracteres especiais
                palavra += caractere.lower()
        if palavra!='':
            todas_as_palavras.append(palavra)
file1.close()
```

```python
print(f'O livro tem {len(todas_as_palavras)} palavras.\n')
```

```python
for i in range(20):
    print(todas_as_palavras[i])
```

```python
palavras = list(set(todas_as_palavras))
```

```python
print(len(palavras))
```

```python
print(palavras)
```

```python
palavras_rpt = {}

for i in todas_as_palavras:
    if i not in palavras_rpt:
        palavras_rpt[i] = 1
    else:
        palavras_rpt[i] += 1

print(palavras_rpt)
# Todas as palavras com a frequência de cada palavra
```

```python
palavras_rpt_ordem = sorted(palavras_rpt, key = palavras_rpt.get, reverse=True)
print(f"{palavras_rpt_ordem}")
```

```python
palavras_frequencia = {}
for i in palavras_rpt_ordem:
    palavras_frequencia[i] = str(round(((palavras_rpt[i])/len(todas_as_palavras))*100,2)) + "%"

for i in range(20):
    print(f"{(list(palavras_frequencia.keys()))[i]} = {palavras_frequencia[(list(palavras_frequencia.keys()))[i]]}")
```
