Desafios e limitações:

    Necessidade de Muitos Dados: Difícil de cumprir em áreas com dados limitados.
    
    Alto Custo Computacional: Requer equipamentos poderosos e caros.
    
    Falta de Transparência: Difícil entender como os modelos tomam decisões.
    
    Dificuldade de Generalização: Modelos podem não se sair bem com dados novos ou diferentes.
    
    Risco de Viés: Modelos podem perpetuar preconceitos dos dados.

Solução apresentada:

    Técnicas Eficientes de Dados e Aprendizado: Uso de menos dados de formas mais inteligentes (como aprendizado por transferência) e geração de dados sintéticos.
    
    Tornar Modelos Mais Claros: Trabalhar em tornar a IA mais explicável.
    
    Melhorar Justiça: Implementar estratégias para reduzir preconceitos nos modelos.

**1. Detecção e Reconhecimento Facial usando CNN**

Aplicação: O reconhecimento facial é uma das aplicações mais prevalentes e impactantes das CNNs, sendo utilizado em uma variedade de contextos, desde a segurança e vigilância até a autenticação pessoal em dispositivos móveis e sistemas bancários. O objetivo é identificar ou verificar a identidade de uma pessoa a partir de uma imagem digital de seu rosto.

Técnica: As CNNs são empregadas para analisar imagens faciais, aprendendo a identificar características únicas e padrões distintos do rosto de cada indivíduo. Modelos como o DeepFace ou FaceNet utilizam arquiteturas profundas para extrair essas características de alta dimensionalidade, que são então comparadas a um banco de dados de rostos conhecidos para identificação ou verificação.

Resultados: Os sistemas de reconhecimento facial baseados em CNN alcançaram níveis de precisão impressionantes, superando até mesmo a capacidade humana em certos benchmarks. Eles são capazes de realizar a identificação facial com alta precisão, mesmo em condições desafiadoras, como variações de iluminação, expressões faciais e ângulos de visão.

Fonte: Um estudo fundamental nesse campo é o artigo "DeepFace: Closing the Gap to Human-Level Performance in Face Verification", publicado no Journal of Machine Learning Research. Este artigo detalha o desenvolvimento do DeepFace, um sistema de reconhecimento facial que reduziu drasticamente a discrepância entre a capacidade de reconhecimento facial humana e a capacidade computacional, atingindo uma precisão de 97,35% em um dataset de referência padrão, o que está muito próximo do desempenho humano estimado em 97,53%.


**2. Previsão de Sequências Temporais com RNNs**

Aplicação: As RNNs são particularmente eficazes para aplicações envolvendo dados sequenciais ou temporais, como a previsão do mercado de ações. Nesse contexto, as RNNs, especialmente as variantes como LSTM (Long Short-Term Memory) ou GRU (Gated Recurrent Units), são utilizadas para modelar e prever movimentos futuros dos preços das ações baseando-se em séries temporais históricas.

Técnica: A RNN utiliza dados históricos do mercado, como preços de fechamento, volume de negociação, e indicadores econômicos, para aprender padrões temporais. A memória de longo prazo das LSTMs permite que o modelo considere dependências de longo prazo nos dados, melhorando a precisão da previsão.

Resultados: A aplicação de RNNs na previsão do mercado de ações demonstrou a capacidade de superar modelos tradicionais e benchmarks de mercado em termos de precisão de previsão, oferecendo insights valiosos para investidores e traders.

Fonte: Um estudo detalhado pode ser encontrado no artigo "A Dual-Stage Attention-Based Recurrent Neural Network for Time Series Prediction", que explora o uso de RNNs com mecanismos de atenção para aprimorar a previsão de séries temporais.

### 4) Vamos fazer uma análise:

a) Utilize a base Animals-10- originalmente do Kaggle (5000 imagens de 10 animais diferentes), mas disponibilizada em uma versão menor aqui;

b) Utilizando um modelo pré-treinado ResNet-50 (presente no Tensorflow/Keras), e realize as extrações de características, fazendo a projeção das imagens no espaço característico (penúltima camada da ResNet-50);

c) Realize a clusterização dos vetores obtidos usando o algoritmo K-médias (K=10);

d) Mostre exemplos de imagens dos 10 clusters obtidos;

e) Os resultados foram satisfatórios? Justifique utilizando seus conhecimentos sobre avaliação de clusters;

f) Disponibilize os códigos, bem como o relatório final no Github;

```python
import numpy as np
import tensorflow as tf
import os
import glob
import matplotlib.pyplot as plt
import collections

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing import image
from sklearn.cluster import KMeans
from tensorflow.keras.applications import ResNet50

# ResNet50 pré-treinado
model = ResNet50(weights='imagenet', include_top=False)
```

```python
# link para a pasta com as imagens
data_dir = 'files/animals10_small'

# Lista de caminhos para todas as imagens
img_paths = []
for extension in ['*.jpg', '*.jpeg', '*.png']:
    img_paths.extend(glob.glob(os.path.join(data_dir, '**/', extension), recursive=True))
```

```python
len(img_paths)
```

```python
# pré processamento das imagens
def img_process(img_path):
    img = image.load_img(img_path, target_size=(448, 448))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.resnet50.preprocess_input(img_array)
```

```python
# Extrair características para cada imagem
features = []
for img_path in img_paths:
    preprocessed_img = img_process(img_path)
    features.append(model.predict(preprocessed_img).flatten())

# Convertendo as características obtidas em matriz
X = np.array(features)

# Algoritmo K-means
kmeans = KMeans(n_clusters=10, random_state=5)
kmeans.fit(X)
```

```python
# Mostrar as imagens dos 10 clusters
for i in range(10):
    cluster_imgs = np.array(img_paths)[np.where(kmeans.labels_ == i)]
    if len(cluster_imgs) >= 5:
        plt.figure(figsize=(10, 5))
        for j in range(min(5, len(cluster_imgs))):
            plt.subplot(1, 5, j+1)
            img = image.load_img(cluster_imgs[j+2])
            plt.imshow(img)
            plt.axis('off')
        plt.suptitle(f'Cluster {i}')
        plt.show()
```

```python
# Avaliação de cada cluster obtido
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X, kmeans.labels_)
print(f'Avaliação dos clusters: Silhueta média = {silhouette_avg}')
```
